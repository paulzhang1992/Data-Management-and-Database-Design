```python
import pandas as pd
import numpy as np
```

Read the information of players' regular totalstats

```python
data_player_regular_totalstats = pd.read_csv('player_regular_totalstats.csv')
data_player_regular_totalstats.head(10)
```

Check the missing of the data

```python
data_player_regular_totalstats.apply(lambda x: sum(x.isnull()),axis=0)
```

Deal with the missing part of 'GS'

```python
miss_GS = np.where(pd.isnull(data_player_regular_totalstats['GS']))
miss_GS
```

Change the missing part of 'GS' into 0

```python
data_player_regular_totalstats.ix[miss_GS[0], 'GS'] = int(0)
data_player_regular_totalstats.ix[miss_GS[0],:]
```

Same way to deal with the 'MIN' part

```python
miss_MIN = np.where(pd.isnull(data_player_regular_totalstats['MIN']))
miss_MIN
```

Change the missing part of 'MIN' into 0

```python
data_player_regular_totalstats.ix[miss_MIN[0], 'MIN'] = int(0)
data_player_regular_totalstats.ix[miss_MIN[0],:]
```

Next, deal with the FG3 part

```python
miss_FG3M = np.where(pd.isnull(data_player_regular_totalstats['FG3M']))
miss_FG3A = np.where(pd.isnull(data_player_regular_totalstats['FG3A']))
miss_FG3_PCT = np.where(pd.isnull(data_player_regular_totalstats['FG3_PCT']))
print(miss_FG3M)
print(miss_FG3A)
print(miss_FG3_PCT)
```

Find that the missing part of 'FG3M' and 'FG3A' are the same

```python
test_equal = miss_FG3M[0] == miss_FG3A[0]
np.sum(test_equal)/len(test_equal)
```

Find that the index of missing part of 'FG3M' is subset of missing part of 'FG3_PCT'

```python
inters_miss_FG3 = set(miss_FG3_PCT[0]) & set(miss_FG3M[0])
inters_miss_FG3 = np.array(list(inters_miss_FG3))
len(inters_miss_FG3) == len(miss_FG3M[0])
```

Change those NaN into 0

```python
data_player_regular_totalstats.ix[miss_FG3A[0], 'FG3A'] = 0
data_player_regular_totalstats.ix[miss_FG3A[0], 'FG3M'] = 0
data_player_regular_totalstats.ix[miss_FG3A[0], 'FG3_PCT'] = 0
data_player_regular_totalstats.ix[miss_FG3A[0], ['FG3A', 'FG3M','FG3_PCT']]
```

Deal with the 'FG3_PCT' part that is not contained in the 'FG3M'

```python
comple_miss_FG3 = set(miss_FG3_PCT[0]) - set(miss_FG3A[0])
comple_miss_FG3 = list(comple_miss_FG3); comple_miss_FG3 = np.array(comple_miss_FG3)
comple_miss_FG3 = sorted(comple_miss_FG3)

data_temp1 = data_player_regular_totalstats.ix[comple_miss_FG3, ['FG3M', 'FG3A', 'FG3_PCT']]
data_temp1
```

There is no missing in 'FG3M' and 'FG3A' in the data_temp1

```python
data_temp1.apply(lambda x: sum(x.isnull()),axis=0)
```

So, we can compute the 'FG3_PCT' according to the 'FG3M' and 'FG3A', and if 'FG3A' equals to 0 we change 'FG3_PCT' into 0

```python
data_player_regular_totalstats.ix[comple_miss_FG3, 'FG3_PCT'] = \
    data_player_regular_totalstats.ix[comple_miss_FG3, 'FG3M'] / data_player_regular_totalstats.ix[comple_miss_FG3, 'FG3A']

for i in comple_miss_FG3:
    if data_player_regular_totalstats.ix[i,'FG3A'] == 0:
        data_player_regular_totalstats.ix[i,'FG3_PCT'] = 0

data_player_regular_totalstats.ix[comple_miss_FG3, ['FG3M', 'FG3A', 'FG3_PCT']]
```

Deal with the 'REB' part

```python
miss_OREB = np.where(pd.isnull(data_player_regular_totalstats['OREB']))
miss_DREB = np.where(pd.isnull(data_player_regular_totalstats['DREB']))
miss_REB = np.where(pd.isnull(data_player_regular_totalstats['REB']))

comple_miss_REB = set(miss_OREB[0]) - set(miss_REB[0])
comple_miss_REB = list(comple_miss_REB); comple_miss_REB = np.array(comple_miss_REB)
comple_miss_REB = sorted(comple_miss_REB)

data_temp2 = data_player_regular_totalstats.ix[comple_miss_REB, ['OREB', 'DREB', 'REB']]
data_temp2
```

```python
data_temp2.apply(lambda x: sum(x.isnull()),axis=0)
```

We change these NaN in 'OREB' into 0 and 'DREB' in to the number of 'REB'

```python
data_player_regular_totalstats.ix[comple_miss_REB, 'OREB'] = 0
data_player_regular_totalstats.ix[comple_miss_REB, 'DREB'] = data_player_regular_totalstats.ix[comple_miss_REB, 'REB']
```

As for those that even not contain the data of 'REB', we define those data as 0

```python
inters_miss_REB = set(miss_OREB[0]) & set(miss_REB[0])
inters_miss_REB = list(inters_miss_REB); inters_miss_REB = np.array(inters_miss_REB)
inters_miss_REB = sorted(inters_miss_REB)

data_temp3 = data_player_regular_totalstats.ix[inters_miss_REB, ['OREB', 'DREB', 'REB']]
data_temp3
```

All data in data_temp3 is NaN

```python
data_temp3.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
data_player_regular_totalstats.ix[inters_miss_REB, 'OREB'] = 0
data_player_regular_totalstats.ix[inters_miss_REB, 'DREB'] = 0
data_player_regular_totalstats.ix[inters_miss_REB, 'REB'] = 0

data_player_regular_totalstats.ix[inters_miss_REB, ['OREB', 'DREB', 'REB']]
```

As for the steal, block and tornover, we just assign the missing data as 0

```python
miss_STL = np.where(pd.isnull(data_player_regular_totalstats['STL']))
miss_BLK = np.where(pd.isnull(data_player_regular_totalstats['BLK']))
miss_TOV = np.where(pd.isnull(data_player_regular_totalstats['TOV']))

np.sum(miss_STL[0] == miss_BLK[0])
```

the number of miss_STL is the same as the intersection of those 'STL' and 'BLK', so, the data in data_temp4 are all NaN

```python
inters_miss_3 = set(miss_STL[0]) & set(miss_BLK[0])
inters_miss_3 = list(inters_miss_3); inters_miss_3 = np.array(inters_miss_3)
inters_miss_3 = sorted(inters_miss_3)

data_temp4 = data_player_regular_totalstats.ix[inters_miss_3, ['STL','BLK']]
data_temp4
```

As for there are all NaN, so we change them into 0

```python
data_temp4.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
data_player_regular_totalstats.ix[miss_STL[0], 'STL'] = 0
data_player_regular_totalstats.ix[miss_BLK[0], 'BLK'] = 0
data_player_regular_totalstats.ix[miss_STL[0], ['STL', 'BLK']]
```

```python
data_temp5 = data_player_regular_totalstats.ix[miss_TOV[0], ['TOV']]
data_temp5
```

Change those NaN into 0

```python
data_player_regular_totalstats.ix[miss_TOV[0], ['TOV']] = 0
data_player_regular_totalstats.ix[miss_TOV[0], ['TOV']]
```

```python
data_player_regular_totalstats.apply(lambda x: sum(x.isnull()),axis=0)
```

After checking the data not contain NaN, we save it

```python
data_player_regular_totalstats.to_csv('revised_player_regular_totalstats.csv', index=False)
```
